```python
def getTweetsTime(df):
    new_time = []
    for i in range(len(df)):
        new_time.append(time.strftime('%Y-%m-%d', time.strptime(df['created_str'].iloc[i],'%a %b %d %H:%M:%S +0000 %Y')))
    df['created_str'] = pd.Series(new_time)
    return df
```

```python
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)
```
